```python
import pandas as pd
import numpy as np
import us
import datetime
```

```python
today = datetime.datetime.now().date()
dates = [today - datetime.timedelta(days=i) for i in range(365*3)]

state_shop = [f'{st.name}-Shop{i}' for st in us.states.STATES for i in range(1, np.random.randint(50,151))]

my_index = pd.MultiIndex.from_product([dates,state_shop],
                          names = ['Date','State_Shop'])

sample_df = pd.DataFrame(index=my_index, columns=['number_of_customer_visits','revenue','profit'])
sample_df
```

```python
sample_df.number_of_customer_visits = np.random.randint(100,1000,[len(sample_df)])
sample_df.revenue = sample_df.number_of_customer_visits * np.round(np.random.random([len(sample_df)]) *13 +2,3)
sample_df.profit = sample_df.revenue * np.round(np.random.random([len(sample_df)]) *0.05 +0.005,3)
sample_df
```

```python
sample_df = sample_df.reset_index()

sample_df = sample_df.join(
    sample_df.State_Shop.str.split('-',expand=True).rename(columns = {0:'State',1:'Location'})
)
```

```python
sample_df.Date = pd.to_datetime(sample_df.Date)

sample_df['Month'] = sample_df.Date.dt.month
sample_df['Year'] = sample_df.Date.dt.year
sample_df
```

```python
sample_df.revenue = sample_df.revenue.astype(float)
sample_df.profit = sample_df.profit.astype(float)

sample_df.revenue = sample_df.revenue.round(2)
sample_df.profit = sample_df.profit.round(2)

sample_df.number_of_customer_visits = sample_df.number_of_customer_visits.astype('int16')
sample_df.Month = sample_df.Month.astype('int16')
sample_df.Year = sample_df.Year.astype('int16')
```
